```python
# Cài đặt thư viện
!pip install pyvi matplotlib

# Import thư viện
from pyvi import ViTokenizer
from collections import Counter
import matplotlib.pyplot as plt

# Đoạn văn bản tiếng Việt
text = """
Chào mừng bạn đến với OpenAI. Chúng tôi là một tổ chức nghiên cứu với mục tiêu phát triển trí tuệ nhân tạo an toàn và có lợi cho mọi người. Hôm nay là một ngày đẹp trời và chúng tôi rất vui được gặp bạn.
"""

# Tách từ
tokenized_text = ViTokenizer.tokenize(text)
print("Đoạn văn bản sau khi tách từ:\n", tokenized_text)

# Đếm tần suất xuất hiện của các từ
word_counts = Counter(tokenized_text.split())
print("\n10 từ phổ biến nhất:\n", word_counts.most_common(10))

# Lấy dữ liệu cho 10 từ phổ biến nhất
common_words = word_counts.most_common(10)
words, counts = zip(*common_words)

#
plt.figure(figsize=(10, 5))
plt.bar(words, counts)
plt.xlabel('Từ')
plt.ylabel('Tần suất')
plt.title('10 từ phổ biến nhất')
plt.xticks(rotation=45)
plt.show()
```
